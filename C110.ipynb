```python
import statistics
import plotly.figure_factory as ff
import pandas as pd
import random
import csv
import plotly.graph_objects as go
df = pd.read_csv("data.csv")
data = df["temp"].to_list()
mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)
SD = statistics.stdev(data)
print("Mean: ",mean)
print("Median: ",median)
print("Mode: ",mode)
print("Standard Deviation: ",SD)
fig = ff.create_distplot([data],["temp"],show_hist=False)
fig.show()
```

```python
ds = []
for i in range(0,100):
  ra = random.randint(0,len(data))
  value = data[ra]
  ds.append(value)
me = statistics.mean(ds)
stdev = statistics.stdev(ds)
print(me)
print(stdev)
fig = ff.create_distplot([ds],["value"],show_hist=False)
fig.show()
```

```python
def rsm(counter):
  dataset = []
  for i in range(0,counter):
    ri = random.randint(0,len(data))
    val = data[ri]
    dataset.append(val)
  mee = statistics.mean(dataset)
  print(mee)
```

```python
def sf(mean_list):
  dataset = mean_list
  fig = ff.create_distplot([dataset],["temp"],show_hist=False)
  fig.show()
```

```python
def checkmean():
  ml = []
  for i in range(0,1000):
    soms = rsm(100)
    ml.append(soms)
    sf(ml)
checkmean()
```
